```python
train_df["title"]=train_df["Name"].str.extract(r',\s*(\w+)\.')
print(train_df["title"])
```

```python
median_age=train_df.groupby("title")["Age"].median()
print(median_age)
```

```python
train_df["Age"]=train_df["Age"].fillna(train_df["title"].map(median_age))
train_df["Age"].isnull().sum()
```

```python
mode=train_df["Embarked"].mode()
print(mode)
```

```python
train_df["Embarked"]=train_df["Embarked"].fillna(mode)
train_df["Embarked"].isnull().sum()
```

```python
train_df["deck"]=train_df["Cabin"].str[0]
train_df["deck"]=train_df["deck"].fillna("U")
print(train_df["deck"])
```

```python
categorical_cols = ['Sex', 'Embarked', 'Pclass', 'title', 'deck']
train_df = pd.get_dummies(train_df, columns=categorical_cols, drop_first=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
columns=['PassengerId','Name','Ticket','Cabin','Survived','Fare','Age',"Age group"]
x=train_df.drop(columns,axis=1)
y=train_df["Survived"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=21)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,y_pred)
print('Model accuracy : ',accuracy)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print("Confusion Matrix is: ",cm)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model=RandomForestClassifier(random_state=21)
rf_model.fit(x_train,y_train)
```

```python
y_pred_rf=rf_model.predict(x_test)
accuracy_rf=accuracy_score(y_test,y_pred_rf)
print(accuracy_rf)
cm_rf=confusion_matrix(y_test,y_pred_rf)
print(cm_rf)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid={
    'n_estimators':[50,100,200,300],
    'max_depth':[None,5,10,20,30],
    'min_samples_split':[2,5,10]
}
grid_search=GridSearchCV(estimator=rf_model,param_grid=param_grid,cv=5,scoring="accuracy")
grid_search.fit(x_train,y_train)
print("Best Parameters:", grid_search.best_params_)
print("Best Accuracy:", grid_search.best_score_)
```
